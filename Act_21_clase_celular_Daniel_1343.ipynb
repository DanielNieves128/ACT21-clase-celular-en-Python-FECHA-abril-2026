```python
class Celular:
  def __init__(self,marca,modelo, precio):
      self.marca = marca
      self.modelo = modelo
      self.precio = precio
      self.encendido = False

  def encender(self):
      self.encendido = True
      print(f"El {self.modelo} se ha encendido.")

  def mostrar_info(self):
      print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")

# Creacion de objetos (instancias)
celular1 = Celular("Samsung", "Galaxy S21", 800)
celular2 = Celular("Apple", "iPhone 13", 900)
# llamadas a las funciones para cada objeto ceclular1 y celular2
print("celular1 :")
celular1.mostrar_info()
celular1.encender()

print("\ncelular2 :")
celular2.mostrar_info()
celular2.encender()

print("programa terminado")
print("Creador de software por Daniel Nieves NL= 41")
```
